## Dropped Alpha Method

```python
import statsmodels.api as sm
x = data[['Dummy_male', 'D_female']]
y = data['Income']

model = sm.OLS(y,x).fit()
predictions = model.predict(x)
model.summary()
```

## Dropped Dummy Method

### Now in second dummy regression model, we will drop one dummy and include the intercept in the regression model. The intercept will take the value of dropped dummy and slope coefficient of included dummy will indicate the differential average.

### -----------------------------------------------------

```python
import statsmodels.api as sm
x = data[['North', 'east', 'South']]
y = data['Income']


model = sm.OLS(y,x).fit()
predictions = model.predict(x)
model.summary()
```

```python
import statsmodels.api as sm
x = data[['east', 'South']]
y = data['Income']

x = sm.add_constant(x)
model = sm.OLS(y,x).fit()
predictions = model.predict(x)
model.summary()
```
